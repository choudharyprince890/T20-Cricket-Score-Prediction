```python
import os
import pandas as pd
import numpy as np
from yaml  import safe_load
from tqdm import tqdm
```

### iterating throught the files in data folder

```python
path_of_files = []
for file in os.listdir('data'):
    path_of_files.append(os.path.join('data',file))
print(path_of_files)
```
